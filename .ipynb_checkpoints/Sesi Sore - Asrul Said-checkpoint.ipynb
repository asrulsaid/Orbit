### Quiz Dukun Seri


```python
import random

game = 0
sum_hum = 0
sum_com = 0
list_angka = [i for i in range(-6, 7)]
while True:
    hum = int(input("Masukan Angka: "))
    if hum < -6 or hum > 6:
        print("Masukkan angka dari rentang -6 hingga 6 !!!")
        continue
        
    com = random.choice(list_angka)
    sum_hum += hum
    sum_com += com
    game += 1
    
    print("Kamu: \t\t\t", hum)
    print("Komputer: \t\t", com)
    print("Jumlah Angka Kamu: \t", sum_hum)
    print("Jumlah Angka Komputer: \t", sum_com)
    print()
    
    if sum_hum == sum_com:
        break
        
print("Anda menang di game ke", game)
```

### Tugas membuat Game menggunakan Looping

```python
import random

com = []
game = 0
count = 0
print("GAME TEBAK PIN")
n = int(input("Masukkan Jumlah Angka yg ingin ditebak: "))
for i in range(n):
    com.append(random.randint(1,10))

while True:
    count += 1
    hum = []
    for i in range(n):
        print("Angka ke", i+1)
        hum.append(int(input()))
        
    if hum == com:
        print()
        print("Selamat !!!")
        print("Kamu menebak sebanyak",count, "kali")
        break
    else:
        print("Salah, Coba Lagi")
        print()
```
