```python
def hitung(order):
    mn = list(menu)
    sub_total = []
    total = 0
    for item in order:
        price = menu[mn[item-1]]
        sub_total.append(order[item] * price)
    return sub_total

def cetak_menu(menu):
    i = 1
    print("====== DAFTAR MENU ======")
    for key in menu:
        print("{}. {} \t Rp.{}".format(i, key,menu[key]))
        i += 1
    print("=" * 25)

def struk(order,sub_total):
    mn = list(menu)
    dash = '=' * 80
    line = '-' * 80
    head = "Menu \t\tJumlah \t\tHarga \t\tSub Total"
    body = ""
    total = 0
    i = 0
    for item in order:
        body += mn[item-1] + "\t"+str(order[item])+ "\t\tRp." + str(menu[mn[item-1]]) + "\tRp."+str(sub_total[i])+"\n"
        total += sub_total[i]
        i += 1
    total = "Total\t\t\t\t\t\tRp."+str(total)
    struk = "{}\n{}\n{}\n{}\n{}\n{}\n{}".format(dash, head, dash, body, line, total,dash)
    return struk

def cetak_struk(nota):
    struk = open("AB_Chicken.txt", "w")
    struk.write(nota)
    struk.close()
    return struk
    
    
menu = {"Ayam Geprek":15000, "Ayam Bakar": 13000, "Ayam Penyet": 14000, "Ayam Goreng": 12000}
order = {}
while True:
    cetak_menu(menu)
    pilihan = int(input("Mau makan apa hari ini? : "))
    qty = int(input("Jumlah : "))
    
    try:
        order[pilihan] = qty + order[pilihan]
    except:
        order[pilihan] = qty
        
    lanjut = input("Mau nambah lagi? (y/n) : ")
    print()
    if lanjut == 'y':
        continue
    elif lanjut == 'n':
        sub_total = hitung(order)
        nota = struk(order, sub_total)
        cetak_struk(nota)
        file = open("AB_Chicken.txt", "r")
        print(file.read())
        break
    else:
        print("Pilihan Salah")
    
```
